```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def kl_divergence(p, q):
    return (p*np.log2(p/q)).sum()
```

### Graphs for 4b and 4c

For derivation of values at $\lambda=1$, please see handwritten homework.

```python
lambdas = np.linspace(0, 1, 100)
D1 = [kl_divergence(np.array([1/3,1/3,1/3]), np.array([l/3,(3-l)/6,(3-l)/6])) for l in lambdas]
D2 = [kl_divergence(np.array([l/3,(3-l)/6,(3-l)/6]), np.array([1/3,1/3,1/3])) for l in lambdas]
D2[0] = np.log2(3)-1
plt.plot(lambdas, D1, label="D(p||q_lambda)")
plt.plot(lambdas, D2, label="D(q_lambda||p)")
plt.title("Divergence vs lambda")
plt.xlabel("Lambda")
plt.ylabel("Bits")
plt.legend()
plt.show()
```
